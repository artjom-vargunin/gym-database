```python
import string    
import random  
import pandas as pd
import numpy as np

def rstr(u,l,n,k): #random string
    return ''.join(random.choices(string.ascii_uppercase*u + string.ascii_lowercase*l + string.digits*n, k = k))
```

## Login table

```python
LoginUsername_l=10  #length for username in table login
LoginPassword_l=20
Login_n=1000  #number of rows in table login

LoginUsernames=[rstr(1,1,1,LoginUsername_l) for i in range(Login_n)]
LoginPasswords=[rstr(1,1,1,LoginPassword_l) for i in range(Login_n)]

Login = pd.DataFrame(data={'Username': LoginUsernames, 'Password': LoginPasswords})
Login.sample(2)
```

## Member table

```python
Member_n=int(np.floor(0.9*Login_n))  # Member_n+Trainer_n=Login_n

MemberID=np.arange(1,Member_n+1)

Login_ind=np.arange(Login_n)
np.random.shuffle(Login_ind)

MemberUsername=np.array(LoginUsernames)[Login_ind[:Member_n]]

MemberEmail_l=10
MemberEmail=[rstr(0,1,1,MemberEmail_l)+'@gmail.com' for i in range(Member_n)]

MemberAddress=[rstr(1,0,0,1)+rstr(0,1,0,k = 5)+' '+rstr(0,0,1,2)+'-'+rstr(0,0,1,2) for i in range(Member_n)]

MemberGender=random.choices([0,1],k=Member_n)
MemberYBirthyear=random.choices(np.arange(1950,2010),k=Member_n) 
MemberYBirth=[MemberYBirthyear[i].astype('str')+'-01-01' for i in range(len(MemberYBirthyear))]  #date=yyyy-mm-dd
MemberFmane_l=5
MemberLmane_l=7
MemberFname=[rstr(1,0,0,1)+rstr(0,1,0,MemberFmane_l) for i in range(Member_n)]
MemberLname=[rstr(1,0,0,1)+rstr(0,1,0,MemberLmane_l) for i in range(Member_n)]

Member = pd.DataFrame(data={'Member_ID':MemberID, 'Username':MemberUsername, 'Email':MemberEmail, 'Address':MemberAddress, 'Gender':MemberGender,'Y_birth':MemberYBirth,'F_name':MemberFname,'L_name':MemberLname})
Member.head(5)
```

## Trainer table

```python
Trainer_n=Login_n-Member_n  # Member_n+Trainer_n=Login_n

TrainerID=np.arange(1,Trainer_n+1)

TrainerUsername=np.array(LoginUsernames)[Login_ind[Member_n:]]

TrainerEmail_l=10
TrainerEmail=[rstr(0,1,1,TrainerEmail_l)+'@gmail.com' for i in range(Trainer_n)]

TrainerAddress=[rstr(1,0,0,1)+rstr(0,1,0,k = 5)+' '+rstr(0,0,1,2)+'-'+rstr(0,0,1,2) for i in range(Trainer_n)]

TrainerGender=random.choices([0,1],k=Trainer_n)
TrainerSalary=random.choices(np.arange(5,50),k=Trainer_n) #tunnitasu

TrainerYbirthyear=random.choices(np.arange(1950,2010),k=Trainer_n)
TrainerYbirth=[TrainerYbirthyear[i].astype('str')+'-01-01' for i in range(len(TrainerYbirthyear))]  #date=yyyy-mm-dd

TrainerAcc=[rstr(0,0,1,9) for i in range(Trainer_n)]
TrainerFname=[rstr(1,0,0,1)+rstr(0,1,0,MemberFmane_l) for i in range(Trainer_n)]
TrainerLname=[rstr(1,0,0,1)+rstr(0,1,0,MemberLmane_l) for i in range(Trainer_n)]

Trainer = pd.DataFrame(data={'Trainer_ID':TrainerID, 'Username':TrainerUsername, 'Email':TrainerEmail, 'Address':TrainerAddress, 'Gender':TrainerGender,'Salary':TrainerSalary,'Y_birth':TrainerYbirth,'Acc_nr':TrainerAcc,'F_name':TrainerFname,'L_name':TrainerLname})
Trainer.head(5)
```

## Time table

```python
Time_n=10000
week='2022-01-0' # we generate data for week starting from this day

TimeID=np.arange(1,Time_n+1)
TimeLogin_comes=random.choices(LoginUsernames,k=int(Time_n/2)) #we generate data for coming/leaving of Time_n/2 members 
TimeLogin=np.append(TimeLogin_comes,TimeLogin_comes)

Timehour=random.choices(np.arange(6,22),k=int(Time_n/2))
Timeday=random.choices(np.arange(1,8),k=int(Time_n/2))
#check how often members come to workouts/ to individual trainings (SQL)

TimeTime_comes=[week+Timeday[i].astype('str')+' '+Timehour[i].astype('str')+':30:00' for i in range(len(Timeday))]  #time = 2020-01-01 00:00:00
TimeTime_leaves=[week+Timeday[i].astype('str')+' '+(Timehour[i]+2).astype('str')+':30:00' for i in range(len(Timeday))]
TimeTime=np.append(TimeTime_comes,TimeTime_leaves)

Time = pd.DataFrame(data={'Time_ID':TimeID,'Username':TimeLogin, 'Time':TimeTime})
Time.loc[Time['Username']=='ALRnjJHxTz'].sort_values(by=['Time'])
```

## Announcements table

```python
Ann_n=50

AnnID=np.arange(1,Ann_n+1)
AnnLogin=random.choices(LoginUsernames,k=Ann_n)
AnnDesc=[rstr(0,1,0,15) for i in range(Ann_n)]
Annday=random.choices(np.arange(1,8),k=Ann_n)
Annhour=random.choices(np.arange(0,24),k=Ann_n)
AnnDate=[week+Annday[i].astype('str')+' '+Annhour[i].astype('str')+':00:00' for i in range(len(Annday))] 

Announcement = pd.DataFrame(data={'Ann_ID':AnnID,'Username':AnnLogin, 'Description':AnnDesc,'Date':AnnDate})
Announcement.head(10)
```

## Rooms table

```python
Room_n=5

RoomID=np.arange(1,Room_n+1)
RoomAvailability=random.choices([0,1],k=Room_n)

RoomName=[rstr(1,0,0,1)+rstr(0,1,0,5) for i in range(Room_n)]

RoomSize=[200,50,50,30,100]
np.random.shuffle(RoomSize)

Room = pd.DataFrame(data={'Room_ID':RoomID,'Name':RoomName,'Size':RoomSize, 'Availability':RoomAvailability})
Room.head(10)
```

## Workouts table

```python
Workout_n=315  #Room_n*timeslots*days=5*9*7=315  
#due to randomness some workouts have to be adjusted due to overlapping times/rooms (SQL)

WorkoutID=np.arange(1,Workout_n+1)

WorkoutTrainerID=random.choices(TrainerID,k=Workout_n)

WorkoutRoom=random.choices(RoomID,k=Workout_n)
WorkoutMaxAttendance=random.choices(np.arange(3,15),k=Workout_n)  
#due to randomness some workouts with large maxAttendance can be in Rooms with small capacity (SQL)
WorkoutType=random.choices(np.arange(1,20),k=Workout_n)
WorkoutStarthour=random.choices(np.arange(10,18,1),k=Workout_n)
WorkoutStartday=random.choices(np.arange(1,8),k=Workout_n)
WorkoutStart=[week+WorkoutStartday[i].astype('str')+' '+WorkoutStarthour[i].astype('str')+':00:00' for i in range(Workout_n)]  #time = 2020-01-01 00:00:00
WorkoutEnd=[week+WorkoutStartday[i].astype('str')+' '+(WorkoutStarthour[i]+1).astype('str')+':00:00' for i in range(Workout_n)]

Workout = pd.DataFrame(data={'Workout_ID':WorkoutID,'Trainer_ID':WorkoutTrainerID, 'Room_ID':WorkoutRoom, 'Max_attendance':WorkoutMaxAttendance,'Type':WorkoutType,'Start_time':WorkoutStart,'End_time':WorkoutEnd})
Workout.sample(5)
```

## Plan table

```python
Plan_n=2835  #Room_n*timeslots*days*mean(max_attendance)=5*9*7*9=2835  
#due to randomness some plans have to be rejected due to full max_attendance (SQL)

PlanID=np.arange(1,Plan_n+1)

PlanMemberID=random.choices(MemberID,k=Plan_n)
#due to ramdomness same members can do same workouts several times (SQL)
PlanWorkoutID=random.choices(WorkoutID,k=Plan_n)

Plan = pd.DataFrame(data={'Plan_ID':PlanID,'Member_ID':PlanMemberID, 'Workout_ID':PlanWorkoutID})
Plan.sample(5)
```

## Membership table

```python
Membership_n=5  
MembershipID=np.arange(1,Membership_n+1)

Membership_ind=np.arange(1,Membership_n+1)
np.random.shuffle(Membership_ind)

MembershipType=Membership_ind
MembershipPrice=Membership_ind*10

Membership = pd.DataFrame(data={'Membership_ID':MembershipID,'Type':MembershipType, 'Price':MembershipPrice})
Membership.head(5)
```

## Receivables table

```python
Rec_n=10000  
RecID=np.arange(1,Rec_n+1)

RecMemberID=random.choices(MemberID,k=Rec_n)
RecMembership=random.choices(MembershipID,k=Rec_n)

Recday=random.choices(np.arange(1,8),k=Rec_n)
RecDate=[week+Recday[i].astype('str')+' 00:00:00' for i in range(Rec_n)]  #time = 2020-01-01 00:00:00
RecReceived=random.choices(MembershipPrice,k=Rec_n)
#due to ramdomness member can pay less/more than required (SQL)
#how many times left till membership is full/expired(SQL)?
Receivables = pd.DataFrame(data={'Received_ID':RecID,'Membership_ID':RecMembership, 'Member_ID':RecMemberID,'Date':RecDate,'Received_amount':RecReceived})
Receivables.head(5)
```
